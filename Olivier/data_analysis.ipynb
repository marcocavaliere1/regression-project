```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df = pd.read_csv('../data/cases_per_age.csv')
```

- The 20/3/2020 lockdown
- The reopening of schools on 31/8/2020
- The 25/12/2020 lockdown

```python
#20/3/2020
df.Date = pd.to_datetime(df.Date)

df_lockdown = df[(df.Date > pd.Timestamp('2020-06-01')) & (df.Date < pd.Timestamp('2020-10-31'))]


fig, ax = plt.subplots(figsize=(6,4))
ax = plt.plot(df_lockdown['Date'], df_lockdown['0-9'], color='r')
ax = plt.plot(df_lockdown['Date'], df_lockdown['10-19'], color='g')
ax = plt.plot(df_lockdown['Date'], df_lockdown['20-29'], color='b')
ax = plt.plot(df_lockdown['Date'], df_lockdown['30-39'], color='orange')
ax = plt.plot(df_lockdown['Date'], df_lockdown['40-49'], color='yellow')

# ax.text(0.3,6500,'Threshold',rotation=90, size=9,
#          ha="center", va="center",
#          bbox=dict(boxstyle="round", ec='k',fc='grey'))
```
